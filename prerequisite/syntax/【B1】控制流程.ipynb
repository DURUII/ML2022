```python
# 猜数字游戏
import random

guess = random.randint(1, 10)

i = 1
while True:  # Python中的布尔值是大写的！
    print("第%d次猜，请输入一个整数数字：" % (i))

    try:  # 异常处理
        temp = int(input())
        i += 1
    except ValueError:
        print("输入无效")
        continue

    if temp == guess:
        print("恭喜你猜对了，就是这个数", guess)  # 依次写入，支持逗号分隔
        break

    elif temp > guess:  # 缩进控制+冒号
        print("大了")

    elif temp < guess:
        print("小了")
```

```python
users = {'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}

# 单元遍历
for idx in users:
    print(idx, users[idx])

print("\n")

# 双元遍历【遍历副本，修改自身】
for key, value in users.copy().items():
    if value == 'inactive':
        del users[key]
    else:
        print(key, value, end=" ")
```

```python
# pass 占位
for i in range(10):
    pass

# 质数查找【for-else循环，无需flag变量】
for n in range(2, 1000):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'equals', x, '*', n // x)
            break
    else:
        # loop fell through without finding a factor
        print(n, 'is a prime number')
```

```python
# 函数【无需定义变量类型、后缀缺省值、注解&解包参数表】
def f(a, b=0):
    print(a + b)
    return None


f(3)
f(3, 4)
f("123", "456")
f(b="456", a="123")

# 函数也可以变量赋值
g = f
g("123", "456")

# 解包
a = ["123", "456"]
g(*a)

a = {'b': "456", 'a': "123"}
g(**a)


# 函数覆盖
def f(a):
    a.append("A")
    return None


# 数组传递引用
a = []
f(a)
print(a)
```
